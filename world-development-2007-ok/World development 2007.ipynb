```python
#!pip install Matplotlib
import matplotlib.pyplot as plt
import csv
import numpy as np 
```

```python
with open('gapminder.csv') as file: 
    reader = csv.reader(file)
    data = list(reader)
print(len(data)) 
```

```python
# getting the useful data 
life_exp = list()
for element in data:
    life_exp.append(element[5])
life_exp.pop(0)
print(len(life_exp))
    
gdp_cap = list()
for element in data: 
    gdp_cap.append(element[6])
gdp_cap.pop(0)
print(len(gdp_cap))

pop = list()
for element in data: 
    pop.append(element[3])
pop.pop(0)
print(len(pop))
```

```python
# transforming datas 
gdp_cap_float = list()
for element in gdp_cap:
    gdp_cap_float.append(float(element))

life_exp_float = list()
for element in life_exp:
    life_exp_float.append(float(element))

pop_float = list()
for element in pop:
    pop_float.append(float(element))

np_pop = np.array(pop_float)
np_pop = np_pop / 1000000
np_pop = np_pop * 2
```

```python
# setting the color 
base_color = {
    'Asia':'red',
    'Europe':'green',
    'Africa':'blue',
    'Americas':'yellow',
    'Oceania':'black'
}

# get the all data 
cont = list()
for element in data:
    cont.append(element[4])
cont.pop(0)
    
# asign color to data
col = list()
for element in cont:
    col.append(base_color[element])
```

```python
# displaying the data
plt.scatter(x=gdp_cap_float, y=life_exp_float, s=np_pop, c=col, alpha=0.5)
plt.xscale('log')
plt.xlabel('GDP per Capita [in UDS]')
plt.ylabel('Life Expectacy [in years]')
plt.xticks([1000, 10000, 100000], ['1K', '10K', '100K'])
plt.yticks([40, 50, 60, 70, 80])
plt.text(30470.0167, 80, 'France')
plt.text(4513.480643, 39.613, 'Swaziland')
plt.grid(True)
plt.savefig('myData.png')
```
